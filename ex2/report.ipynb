### **Enunciado do exercicio:**

2. Construir uma classe Python que implemente o  EcDSA a partir do “standard” FIPS186-5:

    - A implementação deve conter funções para assinar digitalmente e verificar a assinatura.
    - A implementação da classe deve usar  uma das “Twisted Edwards Curves” definidas no standard e escolhida  na iniciação da classe: a curva  “edwards25519” ou “edwards448”.

### **Intrepretação do enunciado:**
É desejado que se implemente uma classe em python que represente um EcDSA.

Um EcDSA é "Elliptic Curve Digital Signature Algorithm" ou seja é desejado que seja implementado um Algoritmo de Assinatura Digital recorrendo a Curvas Elípticas.

É desejado que este algoritmo seja implementado a partir do standard *FIPS186-5*, que especifica o Digital Signature Standard (DSS), ou seja, contém a especificação dos algoritmos que devem ser utilizados para implementar assinaturas digitais.  quais as curvas elipticas que devem ser utilizadas na implementação de um EcDSA.

#### **Analise do FIPS186-5**
Após uma leitura do FIPS186-5 é possivel destacar que os algoritmos de hash utilizados na implementação dos algoritmos de assinatura digital devem utilizar funções de hash aprovadas, algumas delas são o FIPS180 e o FIPS202.




Os parametros utilizados para gerar a assinatura devem seguir o SP 800-89 e/ou SP 800-186.



### **Escolha da curva:**
Escolhi implementar
A implementação escolhida irá utilizar Ed448 pois oferece aproximadamente uma segurança de 224-bits ao invés de 128-bits de segurança do Ed25519.
